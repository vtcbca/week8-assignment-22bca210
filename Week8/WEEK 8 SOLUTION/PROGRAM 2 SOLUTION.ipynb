### B.Export sales table data into sales.csv file.

```python
import sqlite3
import csv
conn = sqlite3.connect("my_database.db")
cursor = conn.cursor()
cursor.execute("SELECT * FROM Sales1")
sales_data = cursor.fetchall()
csv_file = "sales.csv"
with open(csv_file, "w", newline="") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(["sid", "year", "totalsales"])
    csv_writer.writerows(sales_data)
conn.close()
print("Data from the 'Sales' table has been exported to {csv_file}.")
```

### C. Write a python scripts that read the sales.csv file and plot a bar chart that shows totalsales of the year. Also decorate the chart with appropriate title, lables, colours etc.

```python
import pandas as pd
df = pd.read_csv("sales.csv")
plt.figure(figsize=(10, 6))
plt.bar(df['year'], df['totalsales'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')
plt.xticks(rotation=45)
plt.grid(axis='y')
```
